## Check the morphological features

```python
words = F.otype.s('word')

for (feat, feat_etcbc) in (
    ('gn', 'gn_etcbc'),
    ('nu', 'nu_etcbc'),
    ('sp', 'sp_etcbc'),
    ('ps', 'ps_etcbc'),
    ('vs', 'vs_etcbc'),
    ('vt', 'vt_etcbc'),
):
    diffs = collections.Counter()
    Fv = Fs(feat).v
    FEv = Fs(feat_etcbc).v
    for w in words:
        vw = Fv(w)
        vew = FEv(w)
        if vw != vew:
            diffs[(vw or '', vew or '')] += 1
    print(f"{feat} x {feat_etcbc}: {len(diffs)} distinctions:")
    for vx in sorted(diffs):
        n = diffs[vx]
        print(f"\t{vx[0]:<6} x {vx[1]:<6} : {n} nodes")
```
